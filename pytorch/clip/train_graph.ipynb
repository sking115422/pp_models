```python
import re
import matplotlib.pyplot as plt

# Define the step size for plotting (e.g., plot every 10th loss value)
step_size = 1

# List to hold loss values
losses = []

# Updated regular expression to capture the epoch number and loss
pattern = re.compile(r"Epoch (\d+)/30, Loss: ([\d\.]+)")

with open("./checkpoints/training_output.txt", 'r') as file:
    for line in file:
        match = pattern.search(line)
        if match:
            loss = float(match.group(2))
            losses.append(loss)

# Plotting every 'step_size' loss value
plt.figure(figsize=(10, 5))
plt.plot(losses[::step_size], marker='o', linestyle='-')  # Slicing the list with step_size
plt.title('Training Loss Over Time (Every {} Points)'.format(step_size))
plt.xlabel('Iteration (every {}th)'.format(step_size))
plt.ylabel('Loss')
plt.grid(True)
plt.show()
```
