```python
import re
import matplotlib.pyplot as plt
```

```python
with open("./tmp.txt", "r") as f:
    log_data = f.read()
```

```python
pattern = re.compile(r'Epoch (\d+).*?Loss: ([0-9.]+)')

epochs = []
losses = []

# Split log_data into lines if it's a single string, adjust if already a list of lines
lines = log_data.split('\n') if isinstance(log_data, str) else log_data

# Extract data using regex
for line in lines:
    match = pattern.search(line)
    if match:
        epoch = int(match.group(1))
        loss = float(match.group(2))
        epochs.append(epoch)
        losses.append(loss)

# Plotting
plt.figure(figsize=(10, 5))
plt.plot(epochs, losses, marker='o', linestyle='-', color='b')
plt.title('Training Loss Per Epoch')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.grid(True)
plt.show()
```
